```python
# 2, 0
with open("/cluster/projects/radiomics/Temp/sejin/slurm_logs/1556312-aaai_predict.out", 'r') as f:
    out2 = f.readlines()
    
with open("/cluster/projects/radiomics/Temp/sejin/slurm_logs/1556313-aaai_predict.out", 'r') as f:
    out2.extend(f.readlines())
```

```python
results_idx = []
for n, line in enumerate(out2):
    if 'now testing' in line:
        results_idx.append(n)

results_timestamp_dict = {}
for i in results_idx:
    if "FAILED" not in out2[i+1]:
        model_name = out2[i].split("/")[0][-17:]
        timestamp  = out2[i+129].split("]")[-1]
        
        results_timestamp_dict[timestamp.strip()] = model_name
        print(model_name, timestamp)        
```

```python
def name(row):
    try:
        return results_timestamp_dict[row.name]
    except:
        return np.nan
```

```python
results = pd.read_csv("/cluster/home/sejinkim/projects/aaai21_survival_prediction/data/predictions/test_results_new.csv", index_col=0)
results['model'] = results.apply(lambda x: name(x), axis=1)
results
```
